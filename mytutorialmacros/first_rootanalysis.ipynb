```python
import uproot
```

```python
file = uproot.open("downloaded_root_files/ODEO_FEB2025_v0_2J2LMET30_data15_periodD.2J2LMET30.root")
```

```python
print(file.keys())
```

```python
tree = file["analysis"]
print(tree.keys())
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lep_pt_data = tree["lep_pt"].array(library="np")
print(type(lep_pt_data))
print(np.size(lep_pt_data))
print(type(lep_pt_data[0]))
# flatten in case I have vectors per event
if isinstance(lep_pt_data[0], (np.ndarray, list)):
    lep_pt_data = np.concatenate(lep_pt_data)
plt.
```

```python
# plot histogram
plt.hist(lep_pt_data,bins=100,histtype="step",color="red",label="lep pt data")
plt.xlabel("Lepton $p_T$ (GeV)")
plt.ylabel("Events")
plt.title("First uproot plot")
plt.legend()
```
